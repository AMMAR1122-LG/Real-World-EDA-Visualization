## ðŸ› ï¸ Managing Outliers

```python
def cap_outliers_iqr(df):
    df_capped = df.copy()
    for col in df_capped.select_dtypes(include=np.number).columns:
        Q1 = df_capped[col].quantile(0.25)
        Q3 = df_capped[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_capped[col] = np.where(df_capped[col] < lower_bound, lower_bound, df_capped[col])
        df_capped[col] = np.where(df_capped[col] > upper_bound, upper_bound, df_capped[col])
    return df_capped


df_capped = cap_outliers_iqr(df)
```

```python
# Choose a column with known outliers
col = 'marginal_adhesion'  # or any column you're interested in

# Visual comparison
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x=df[col])
plt.title(f"Original: {col}")

plt.subplot(1, 2, 2)
sns.boxplot(x=df_capped[col])
plt.title(f"Capped: {col}")

plt.tight_layout()
plt.show()
```
